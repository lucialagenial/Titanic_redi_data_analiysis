### Titanic Dataset

Today we will explore a new dataset, the infamous Titanic dataset. It contains informations on passengers of the Titanic. At kaggle many people already investigated this dataset. Check it out to get new ideas: -> https://www.kaggle.com/c/titanic/overview


The dataset contains 12 columns.

### 2. What is the average fare per class? On which range do the fares per class spread?

```python
# Average fare per class: 
df.groupby(['Pclass'])[['Fare']].mean().reset_index()

```

```python
max_pclass_fare = df.groupby(['Pclass'])[['Fare']].max().round(1).reset_index()
max_pclass_fare.rename(columns={'Pclass':'Pclass', 'Fare':'Max Fare'}, inplace =True)
min_pclass_fare = df.groupby(['Pclass'])[['Fare']].min().round(1).reset_index()
min_pclass_fare.rename(columns= {'Pclass':'Remove', 'Fare':'Min Fare'}, inplace = True)
max_min_pclass_fare = pd.concat([max_pclass_fare, min_pclass_fare], axis=1)

max_min_pclass_fare.drop(columns = 'Remove', axis=1, inplace=True)
max_min_pclass_fare
max_min_pclass_fare['Range'] = (max_min_pclass_fare['Max Fare'] - max_min_pclass_fare['Min Fare'])
max_min_pclass_fare
```

### 5. What is the most frequent last name?


```python
# This option does not seem to work
df['Name'].value_counts().idxmax()
```

```python
import nltk
import re
import string
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize

stop_words = stopwords.words()

def cleaning(text):        
    # converting to lowercase, removing Mr & Miss etc
    text = re.sub('()', '', text)
    text = re.sub('Mr', '', text)
    text = re.sub('Miss', '', text)
    text = re.sub('Mrs', '', text)
    text = re.sub('Master', '', text)
    text = re.sub('Dr', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
 

    # removing the stop-words          
    text_tokens = word_tokenize(text)
    tokens_without_sw = [word for word in text_tokens if not word in stop_words]
    filtered_sentence = (" ").join(tokens_without_sw)
    text = filtered_sentence
    
    return text

dt = df['Name'].apply(cleaning)

from collections import Counter
p = Counter(" ".join(dt).split()).most_common(42)
rslt = pd.DataFrame(p, columns=['Word', 'Frequency'])
print(rslt)
```

```python
# Johnson seems to be the most frequent last name 
for row in dt:
  if 'Johnson' in row:
    print(row)
```
